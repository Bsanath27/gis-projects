# 🎾 Tennis Court Satellite Detector — India

**Pixel-level detection of tennis courts across India using satellite imagery.**

### Pipeline
```
Satellite tiles (Esri World Imagery, free)
        ↓
Stage 1: OpenCV HSV pixel masking + shape validation
        ↓
Stage 2: MobileNetV2 CNN second-stage classifier
        ↓
IoU-based bounding box deduplication
        ↓
GeoJSON + CSV + Folium interactive map
```

### Detected surface types
| Surface | HSV range | Example |
|---|---|---|
| Hard (teal) | H82–100 | Most Indian city courts |
| Hard (green) | H40–82 | DDA, club courts |
| Hard (blue)  | H100–130 | International venues |
| Clay | H5–20 | Red clay, some club courts |
| Grass | H35–75 | Rare in India |
| Synthetic | H55–95 | Increasingly common |

---
**No Overpass API. No OSM. Pure satellite pixel analysis.**

## 3. HSV Threshold Tuning (Visual Tool)

````python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from tile_fetcher import TileFetcher, lat_lon_to_tile
from config import COURT_HSV_RANGES, WHITE_LINE_HSV

fetcher = TileFetcher()

def visualise_hsv_masks(lat: float, lon: float, zoom: int = 18):
    """
    Fetch a tile and show all HSV masks side by side.
    Good for tuning thresholds.
    """
    from tile_fetcher import lat_lon_to_tile
    x, y = lat_lon_to_tile(lat, lon, zoom)
    bgr = fetcher.fetch_bgr(x, y, zoom)
    if bgr is None:
        print(f'Tile fetch failed for {lat},{lon}')
        return

    hsv = cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV)
    rgb = bgr[:, :, ::-1]

    n_surfaces = len(COURT_HSV_RANGES) + 1  # +1 for white lines
    fig, axes = plt.subplots(2, (n_surfaces + 2) // 2 + 1, figsize=(18, 8))
    axes = axes.flatten()

    # Original
    axes[0].imshow(rgb)
    axes[0].set_title('Original', fontweight='bold')
    axes[0].axis('off')

    # Each surface mask
    for i, (name, params) in enumerate(COURT_HSV_RANGES.items(), 1):
        lower = np.array(params['lower'], dtype=np.uint8)
        upper = np.array(params['upper'], dtype=np.uint8)
        mask = cv2.inRange(hsv, lower, upper)
        pct = mask.sum() / 255 / (256*256) * 100
        axes[i].imshow(mask, cmap='Greens')
        axes[i].set_title(f"{params['label']}\n{pct:.1f}% pixels", fontsize=8)
        axes[i].axis('off')

    # White lines
    wl = cv2.inRange(hsv,
                     np.array(WHITE_LINE_HSV['lower'], dtype=np.uint8),
                     np.array(WHITE_LINE_HSV['upper'], dtype=np.uint8))
    wl_pct = wl.sum() / 255 / (256*256) * 100
    axes[len(COURT_HSV_RANGES)+1].imshow(wl, cmap='Blues')
    axes[len(COURT_HSV_RANGES)+1].set_title(f'White lines\n{wl_pct:.1f}% pixels', fontsize=8)
    axes[len(COURT_HSV_RANGES)+1].axis('off')

    # Hide unused axes
    for j in range(len(COURT_HSV_RANGES)+2, len(axes)):
        axes[j].axis('off')

    plt.suptitle(f'HSV masks @ {lat:.4f},{lon:.4f} Z{zoom}', fontsize=12)
    plt.tight_layout()
    plt.show()

# Test on a known court location (DLF Club, Gurugram — hard court)
# Replace with any coordinates near a court you know about
visualise_hsv_masks(28.459, 77.063, zoom=18)
````

## 4. CV Detector — Single Tile Test

````python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
from tile_fetcher import TileFetcher, lat_lon_to_tile
from cv_detector  import CVCourtDetector

fetcher  = TileFetcher()
detector = CVCourtDetector(zoom=18, debug=True)

def test_cv_on_tile(lat: float, lon: float):
    x, y = lat_lon_to_tile(lat, lon, 18)
    bgr  = fetcher.fetch_bgr(x, y, 18)
    if bgr is None:
        print('Tile fetch failed')
        return

    dets = detector.detect(bgr, x, y)
    rgb  = bgr[:, :, ::-1]

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Original
    axes[0].imshow(rgb)
    axes[0].set_title(f'Satellite tile @ {lat:.4f},{lon:.4f}', fontweight='bold')
    for d in dets:
        rect = patches.Rectangle(
            (d.px_x, d.px_y), d.px_w, d.px_h,
            linewidth=2, edgecolor='lime', facecolor='none'
        )
        axes[0].add_patch(rect)
        axes[0].text(d.px_x, d.px_y - 3,
                     f'{d.surface_type} {d.cv_confidence:.2f}',
                     color='lime', fontsize=7, fontweight='bold')
    axes[0].axis('off')

    # Debug overlay
    if dets and dets[0].debug_overlay is not None:
        axes[1].imshow(dets[0].debug_overlay[:, :, ::-1])
        axes[1].set_title('CV detection overlay', fontweight='bold')
    else:
        axes[1].imshow(rgb)
        axes[1].set_title(f'No detections (n={len(dets)})')
    axes[1].axis('off')

    if dets:
        print(f'\n{len(dets)} detection(s):')
        for d in dets:
            print(f'  surface={d.surface_type}  cv_conf={d.cv_confidence:.3f}'
                  f'  area={d.area_m2:.0f}m²  aspect={d.aspect_ratio:.2f}')
    else:
        print('No courts detected in this tile.')

    plt.tight_layout()
    plt.show()

# Test on a tile — replace with coordinates near a known court
test_cv_on_tile(12.934, 77.627)  # Bengaluru Koramangala
````

## 6. Run the Full Scan

````python
# ─── SCAN CONFIGURATION ───────────────────────────────────────────────────────
# Change these before running:

SCAN_MODE  = 'cities'     # 'cities' | 'city' | 'bbox'
SCAN_CITY  = 'Mumbai'     # Only used when SCAN_MODE='city'
SCAN_BBOX  = {            # Only used when SCAN_MODE='bbox'
    'north': 19.28,
    'south': 18.87,
    'east':  73.10,
    'west':  72.70,
}
RESUME_FROM_CHECKPOINT = True   # Set False to restart from scratch

print(f'Mode: {SCAN_MODE}')
if SCAN_MODE == 'cities':
    from tile_fetcher import get_city_tiles
    tiles = get_city_tiles()
    print(f'Tiles to scan: {len(tiles):,}')
    print(f'Cities: {[c[0] for c in __import__("config").INDIA_CITIES]}')
````

````python
import threading
from tqdm.notebook import tqdm
from scanner import run_scan

# Progress bar
pbar = tqdm(total=100, desc='Scanning', unit='%')
last_pct = [0]

def progress_cb(done, total, detections):
    pct = int(done / total * 100) if total > 0 else 0
    delta = pct - last_pct[0]
    if delta > 0:
        pbar.update(delta)
        last_pct[0] = pct
    pbar.set_postfix({'tiles': done, 'courts': detections})

stop_event = threading.Event()

# ─── LAUNCH SCAN ─────────────────────────────────────────────────────────────
courts = run_scan(
    mode=SCAN_MODE,
    city_name=SCAN_CITY if SCAN_MODE == 'city' else None,
    bbox=SCAN_BBOX       if SCAN_MODE == 'bbox' else None,
    resume=RESUME_FROM_CHECKPOINT,
    progress_cb=progress_cb,
    stop_event=stop_event,
)

pbar.close()
print(f'\n✅ Scan complete. {len(courts)} courts found.')
````

````python
# Stop the scan early if needed (run this cell)
stop_event.set()
print('Stop signal sent. Scan will finish current batch then exit.')
````

## 11. Advanced — Fine-Tune CNN on Your Own Data

If you want higher precision, collect satellite tile crops (256×256 PNG/JPG) and organize them:

```
data/
  positive_crops/   ← tiles that contain a tennis court
    court_001.png
    court_002.png
    ...
  negative_crops/   ← tiles that do NOT contain a court
    neg_001.png
    neg_002.png
    ...
```

~200+ images per class is enough for good accuracy.

````python
# Save crops from detections for positive examples
import cv2
from pathlib import Path
from tile_fetcher import TileFetcher, TileCache

def save_detection_crops(courts, n_max=200, out_dir='data/positive_crops'):
    Path(out_dir).mkdir(parents=True, exist_ok=True)
    fetcher = TileFetcher()
    saved = 0
    for court in courts[:n_max]:
        bgr = fetcher.fetch_bgr(court.tile_x, court.tile_y, court.zoom)
        if bgr is not None:
            fname = f'{out_dir}/court_{court.tile_x}_{court.tile_y}.png'
            cv2.imwrite(fname, bgr)
            saved += 1
    print(f'Saved {saved} positive crops to {out_dir}')

# Uncomment to save:
# save_detection_crops(courts)
````
