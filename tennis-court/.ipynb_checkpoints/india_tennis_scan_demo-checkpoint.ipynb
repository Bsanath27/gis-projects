# India Tennis Court Scan Demo

Satellite-only tennis court discovery for India without Overpass. The pipeline scrapes imagery tiles, proposes court candidates with pixel-level GIS heuristics, filters them with a MobileNet checkpoint when available, deduplicates with IoU, and exports GeoJSON plus a Folium map.

## Runtime notes

- Full-country `z18` scraping is too large for a single notebook pass. This notebook defaults to a bounded ROI while keeping the code resumable for longer runs.
- Set `SATELLITE_TILE_TEMPLATE` to an XYZ/WMTS template that you are allowed to scrape. The code intentionally does not hard-code a legally ambiguous default.
- If `labs/gis_tennis/models/mobilenet_tennis_courts.pt` is absent, the scanner falls back to heuristics only.

```python
import os
from pathlib import Path

from IPython.display import IFrame, display

from labs.gis_tennis.pipeline import (
    ScanConfig,
    aoi_from_bbox,
    default_provider,
    estimate_tile_count,
    load_or_download_india_boundary,
    scan_aoi,
    train_mobilenet_binary,
)

ARTIFACTS = Path("labs/gis_tennis/artifacts")
ARTIFACTS.mkdir(parents=True, exist_ok=True)
```

```python
tile_template = os.environ.get("SATELLITE_TILE_TEMPLATE")
if not tile_template:
    raise RuntimeError(
        "Set SATELLITE_TILE_TEMPLATE before running. Example: export SATELLITE_TILE_TEMPLATE='https://.../{z}/{y}/{x}.png'"
    )

provider = default_provider(tile_template)
provider.min_delay_seconds = 0.5
provider.max_retries = 5
provider.timeout_seconds = 40.0
provider.headers = {"User-Agent": "mujoco-gis-tennis-demo/0.1"}
provider
```

```python
india_aoi = load_or_download_india_boundary(ARTIFACTS / "india_adm0.geojson", simplify_tolerance=0.01)
countrywide_tile_estimate = estimate_tile_count(india_aoi, zoom=18)
countrywide_tile_estimate
```

```python
# Demo ROI: Bengaluru urban region. Replace this with india_aoi for large resumable runs.
demo_aoi = aoi_from_bbox(77.45, 12.84, 77.74, 13.12)

config = ScanConfig(
    zoom=18,
    tile_budget=750,
    max_tiles_per_run=250,
    min_heuristic_score=0.58,
    min_classifier_score=0.80,
    iou_threshold=0.35,
)
config
```

```python
result = scan_aoi(demo_aoi, provider=provider, config=config)
result["metadata"]
```

```python
display(IFrame(src=str(result["map_path"]), width="100%", height=700))
print(result["geojson_path"])
print(result["map_path"])
```
